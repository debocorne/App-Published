```python
# 1. Instalá la librería oficial de Groq (si no lo hiciste ya)
!pip install -q groq

# 2. Librerías necesarias
from groq import Groq
import os
from getpass import getpass
from typing import List, Dict

# 3. Ingresar la API Key de Groq
api_key = getpass("Ingresá tu Groq API Key: ")
client = Groq(api_key=api_key)

# 4. Función para consultar al modelo
def generate_response(messages: List[Dict]) -> str:
    """Call LLM to get response"""
    response = client.chat.completions.create(
        model="compound-beta",
        messages=messages,
        max_tokens=1024
    )
    return response.choices[0].message.content

# 5. Ejemplo de mensajes
import json

code_spec = {
    'name': 'intercambiar_claves_valores',
    'description': 'Intercambia las claves y los valores en un diccionario dado.',
    'params': {
        'd': 'Un diccionario con valores únicos.'
    },
}

messages = [
    {
        "role": "system",
        "content": "Sos un ingeniero de software experto que escribe código funcional limpio. Siempre documentás tus funciones."
    },
    {
        "role": "user",
        "content": f"Por favor implementá: {json.dumps(code_spec)}"
    }
]

response = generate_response(messages)
print(response)
```
