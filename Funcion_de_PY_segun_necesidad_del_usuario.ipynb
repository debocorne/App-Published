````python
# 1. Instalar LiteLLM
!pip install -q litellm

# 2. Importar librerías necesarias
import os
from getpass import getpass
import litellm
from typing import List, Dict
import re

# 3. Ingreso de API Key de Groq
api_key = getpass("Ingresá tu Groq API Key: ")
os.environ["GROQ_API_KEY"] = api_key

# 4. Configurar LiteLLM para Groq
litellm.api_key = os.environ["GROQ_API_KEY"]
base_url = "https://api.groq.com/openai/v1"

# 5. Pedir al usuario qué función desea crear
descripcion_funcion = input("¿Qué función en Python te gustaría crear? Describila brevemente: ")

# 6. Crear mensajes para el modelo
messages = [
    {
        "role": "system",
        "content": "Sos un ingeniero de software experto. Respondé con una función en Python que cumpla lo solicitado. Podés incluir una breve explicación luego del código."
    },
    {
        "role": "user",
        "content": f"Por favor, escribí una función en Python que haga lo siguiente: {descripcion_funcion}"
    }
]

# 7. Ejecutar la consulta al modelo
response = litellm.completion(
    model="compound-beta",
    messages=messages,
    max_tokens=1024,
    api_base=base_url,
    api_key=os.environ["GROQ_API_KEY"]
)

respuesta_llm = response["choices"][0]["message"]["content"]

# 8. Mostrar la respuesta completa
print("\nRespuesta completa del modelo:\n")
print(respuesta_llm)

# 9. Extraer el bloque de código
codigo_match = re.search(r"```(?:python)?\s*(.*?)```", respuesta_llm, re.DOTALL)
if codigo_match:
    codigo = codigo_match.group(1).strip()
else:
    lineas = respuesta_llm.splitlines()
    codigo = "\n".join([l for l in lineas if l.strip().startswith("def ") or l.strip().startswith("    ")])

comentario = respuesta_llm.replace(codigo, "").strip()

# 10. Mostrar resultados separados
print("\nCódigo extraído:\n")
print(codigo)

print("\nComentario adicional:\n")
print(comentario)

````
